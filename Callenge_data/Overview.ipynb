### 1 데이터 불러오기
자료 Table 가져오기

```python
# Reviewing Tag info
import os, glob
filelist = glob.glob('./Csv/*.*')
filelist
```

```python
import pandas as pd
df_tag = pd.read_excel(filelist[2], index_col=None)
df_tag.iloc[:34,[0,2,5,6]]
```

```python
df_tag.iloc[34:,[0,2,5,6]]
```

```python
df_org = pd.read_csv(filelist[3], encoding='ms949')
df_org = df_org.drop('CUST_ID', axis=1)  # 불필요한 ID는 제거
print(df_org.head(2))
```

```python
result = {}
for i, tag  in  enumerate(df_tag.변수영문명):
    result[tag] = df_tag.변수명[i]
# 요약 색인으로 변경
df = df_org.rename(columns=result)
df.columns 
```

### 2 상위 columns 첨부하기
데이터 제공기관 분류

```python
top = {}
for i, tag  in  enumerate(df_tag.변수명):
    top[tag] = df_tag.No[i]
cols = [top[col]  for col in df.columns]   
idx = df.columns.tolist()
top = [top[ids]  for ids in idx] 
```

```python
# [ 상위 컬럼 list, 하단 컬럼 list ] (2단 계층컬럼을 삽입)
df.columns = [top , df.columns]  
df = df.sortlevel(0, axis=1)
df.head(2)
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
df_cor = df.corr()
df_cor = df_cor[abs(df_cor) >= 0.3]
plt.subplots(1,1, figsize=(9,7))
matplotlib.rc('font', family='NanumGothic');
sns.heatmap(df_cor);plt.tight_layout(); plt.show()
```

### 4 Hanhwa 컬럼의 영역 살펴보기
보험데이터 살펴보기

```python
Column_index = 'Hanhwa'
print(df[Column_index].columns)
df[Column_index].head(2)
```

```python
# 상관성이 -+ 0.5 이상인 묶음 출력
df_cor = df[Column_index].corr()
df_cor = df_cor.fillna(0)
result = []
for colname in df_cor.columns:
    for i, col in enumerate(df_cor[colname]):
        if col >= 0.5 and col != 1: result.append([colname, df_cor.index[i], float(str(col)[:5])])
# 중복된 내용 제거한 순수한 결과값만 추출
summary , check = [], []
for col in result:
    if col[:2] not in check:
        check.append(col[:2])
        check.append(col[:2][::-1])
        summary.append(col)
    else:
        pass
summary = pd.DataFrame(summary ,columns=['factor1', 'factor2','corr'])
summary = summary.sort_values('corr',ascending=False)
summary.reset_index(inplace=True); summary = summary.iloc[:,1:]; summary
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
df_cor = df_cor[abs(df_cor) >= 0.5]
plt.subplots(1,1, figsize=(10,7))
matplotlib.rc('font', family='NanumGothic');
sns.heatmap(df_cor);plt.tight_layout(); plt.show()
```

```python
# 한화 개별 컬럼의 데이터 살펴보기
# '막내나이'데이터만 분실할뿐 다른 영역은 나름 알차게 정리되어 있다.
print(len(df[Column_index].columns), 'total column counts')
for col in df[Column_index].columns:
    print(col , '\n', set(df[Column_index][col]), '\n\n')
#     print(col , '\n', set(df['Hanhwa'][col].dropna()), '\n\n')
```
