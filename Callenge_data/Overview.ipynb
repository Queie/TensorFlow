### 1 데이터 불러오기
자료 Table 가져오기

```python
# Reviewing Tag info
import os, glob
filelist = glob.glob('./Csv/*.*')
filelist
```

```python
import pandas as pd
df_tag = pd.read_excel(filelist[2], index_col=None)
df_tag.head(2)
```

```python
df_org = pd.read_csv(filelist[3], encoding='ms949')
df_org = df_org.drop('CUST_ID', axis=1)  # 불필요한 ID는 제거
print(df_org.head(2))
```

```python
result = {}
for i, tag  in  enumerate(df_tag.변수영문명):
    result[tag] = df_tag.변수명[i]
# 요약 색인으로 변경
df = df_org.rename(columns=result)
df.columns 
```

### 2 상위 columns 첨부하기
데이터 제공기관 분류

```python
top = {}
for i, tag  in  enumerate(df_tag.변수명):
    top[tag] = df_tag.No[i]
cols = [top[col]  for col in df.columns]   
idx = df.columns.tolist()
top = [top[ids]  for ids in idx] 
```

```python
df.columns = [top , df.columns]
df = df.sortlevel(0, axis=1)
df.head(2)
```

### 3 Data 컬럼의 영역 살펴보기
표준이 되는 기본데이터로 \* 표시는 뭐꼬???

```python
# Data 개별 컬럼의 데이터 살펴보기
print(len(df['Data'].columns), 'total column counts')
for col in df['Data'].columns:
    print(col , '\n', set(df['Data'][col].dropna()), '\n\n')
```

```python
# 성별데이터 0.4 % 의 이물질이 끼어 있다.
sex_estro = [i  for i,check in enumerate(df['Data'].성별)  if check == '*']
len(sex_estro)/ len(df.index) * 100
```

```python
# 연령 데이터의 0.43 % 의 이물질이 끼어 있다.
age_estro = [i  for i,check in enumerate(df['Data'].연령)  if check == '*']
len(age_estro)/ len(df.index) * 100
```

```python
# [err for idx in sex_estro  if idx not in age_estro]
# [err for idx in age_estro  if idx not in sex_estro]
print(age_estro == sex_estro)
# '*' 표시 데이터 살펴보기 
# 총 430개 정보에 아래와 같이 연체여부만 존재
df['Data'].iloc[age_estro,:].head(3) 
```
